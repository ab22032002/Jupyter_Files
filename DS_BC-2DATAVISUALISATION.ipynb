```python
#  sharign the axes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/Standard_Metropolitan_Areas_Data-data.csv")
plt.subplot(1,2,1)
plt.plot(x.work_force,x.income,"--g^")
plt.title("income vs work force")
plt.subplot(1,2,2)
plt.plot(x.hospital_beds,x.income,"--b*")
plt.title('income vs Hospitalbeds')
plt.suptitle("SUB plots")
plt.show()
# sharey =True;
```

```python
fig, ax= plt.subplots(nrows=2,ncols=2,figsize=(6,6),sharey=True, sharex=True)
ax[0,0].plot(x.work_force,x.income,"go")
ax[0,1].plot(x.work_force,x.income,"ro")
ax[1,0].plot(x.work_force,x.income,"bo")
ax[1,1].plot(x.work_force,x.income,"yo")
```
