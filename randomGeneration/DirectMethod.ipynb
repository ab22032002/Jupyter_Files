```python
import string    
import random # define the random module  
S = 10  # number of characters in the string.  
# call random.choices() string module to find the string in Uppercase + numeric data.  
ran = ''.join(random.choices(string.ascii_uppercase + string.digits, k = S))    
print("The randomly generated string is : " + str(ran)) # print the random data  
```

```python
def upcheck(s):
    for i in s:
        if i.isupper():
            return True
```

```python
import string
import random
s=6
for i in range(100):
    ran=''.join(random.choices(string.ascii_uppercase+string.digits+string.ascii_lowercase,k=s))
    if upcheck(ran):
        print(ran)
#     print(ran)
```

```python
def lowcheck(s):
    for i in s:
        if i.islower():
            return True
```

```python
def digcheck(s):
    for i in s:
        if i.isdigit():
            return True
```

```python
import string
import random
s=6
for i in range(100):
    ran=''.join(random.choices(string.ascii_uppercase+string.digits+string.ascii_lowercase,k=s))
    if lowcheck(ran):
        print(ran)
```

```python
import string
import random
s=6
for i in range(100):
    ran=''.join(random.choices(string.ascii_uppercase+string.digits+string.ascii_lowercase,k=s))
    if digcheck(ran):
        print(ran)
```

```python
import string
import random
s=6
for i in range(100):
    ran=''.join(random.choices(string.ascii_uppercase+string.digits+string.ascii_lowercase,k=s))
    if upcheck(ran) and lowcheck(ran) and digcheck(ran):
        print(ran)
```
